```python
def fake_survival(data, column , st, days):
    act_label = label.merge(data, on='acc_id', how='inner')

    fake_survived = act_label[(act_label['survival_time']<= st )]
    k = fake_survived.groupby([column]).count()
    k1 = k[k['day'] <= days ]

    k2 = pd.Series(k1.index, name=column)
    fake_survived = fake_survived.merge(k2, on=column, how='inner')
    return fake_survived

k = fake_survival(act, 'char_id', 1, 1)
k
# act_label.drop(act_label[(act_label['survival_time']==1)&(act_label['day']==1)])


# act[act['acc_id']==59687]
# data = act[act['day']==29]

# for i in range(len(k1)):
#     data.loc[i] = real_suvived[real_suvived['acc_id']==k1.index[i]]
# print(data)
```
